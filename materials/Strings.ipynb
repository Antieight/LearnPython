#### Strings

Переменные бывают типа **str** - строки. *(Чтобы узнать тип переменной всегда можно использовать команду **type**)*

Строка - набор символов, заключенных в кавычки. Для получения длины строки есть команда **len()**

Получить строку от пользователя можно командой **input()**. <br />
Напечатать строку (и не только) - командой **print()**.

```python
greetings = 'Hello, world!'
other_string = input()

print()
print('There can go a string')
print(greetings)
print(type(other_string))
print(len(other_string))
print(other_string)
```

## Обращение по индексу и срезы.

Строка состоит из символов и можно получить любой символ данной строки с помощью оператора квадратных скобок. Можно даже несколько. <br />
Синтаксис такой: <br />
1. **my_string[i]** - **i**-ый символ. При указании отрицательного числа отсчет начинается с конца строки.<br />
2. **my_string[i:j]** - строка, составленная из символов с i по j, включая начальный символ и не включая конечный. <br /> Если опустить индекс **i**, строка не обрежется слева, если **j** - то справа.
3. **my_string[i:j:k]** - как в п.2, **k** указывает скорость движения. По умолчанию **1**. Скорость бывает отрицательной, но в этом случае необходимо правильно выбрать i, j<br />

Первая операция называется обращением по индексу, 2-3  -- срезами. По английски - slice.

### Индексы начинаются с 0!

```python
s = 'Simple is better than complex'
print(s)
print()

# printing second char and the last one.
print('Second character is \t\'' + s[2] + '\'')
print('last one is \t\t\'' + s[-1] + '\'')

print('s[0:10]: \t\t\'' + s[0:10] + '\'')
print('s[:8]: \t\t\t\'' + s[:8] + '\'')
print('s[-7:-2]: \t\t\'' + s[-7:-2] + '\'')
print('s[-4:]: \t\t\'' + s[-4:] + '\'')  # feel the difference with s[-4] !
print('s[9:-1]: \t\t\'' + s[10:-1] + '\'')

print('s[::2] = \t\t ' + s[::2])
print('s[1::2] = \t\t ' + s[1::2])
print('s[::-1] = \t\t ' + s[::-1])
print('s[5::-1] = \t\t ' + s[5::-1])
```

## Функции поиска по строке. find, rfind, count.
1. Чтобы узнать входит ли данный символ (или строка) в нашу строку можно использовать ключевое слово **in**. <br />
 *'ab' in 'abcdefdef' = True*
1. Чтобы найти наименьший индекс, по которому в строке можно обнаружить символ **ch**, нужно использовать функцию **find**: *my_string.find(ch)*. <br />
   Эта функция работает и для подстрок (см. примеры)
2. Если интересует не первое вхождение, а последнее, используем **rfind**.
4. Для подсчета количества вхождений какого-либо символа будем использовать функцию **count**. В случае поиска колличества вхождений других строк есть тонкость, см. примеры.

```python
s = 'abcdefdefaaa'
print("'ab' in 'abcdefdefaaa' is " + str('ab' in s))
print("'ac' in 'abcdefdefaaa' is " + str('ac' in s))
print()

print('searching \'d\' from the left side: ' + str(s.find('d')))
print('searching \'d\' from the right side: ' + str(s.rfind('d')))
print()

print('counting a in ' + s + ': ' + str(s.count('a')))
print('counting aa in ' + s + ': ' + str(s.count('aa')))
print('(last one worked somewhat wrong)')
```

## Функции замены и форматирования строки. replace, format.

1. Чтобы заменить все подстроки ch_old на подстроки ch_new, нужно использовать функцию **replace(old_char, new_char)**
2. Чтобы избегать длинных выражений со сложением строк и использования функции **str()** (в примерах выше она была необходима, так как нельзя складывать строки с нестроками) используют функцию **format**. Принцип ее работы заключается в заполнении строки, содержащей шаблон. Шаблон - строка с выражениями {0}, {1}, ... Функция используется следующим образом: **pattern_string.format(param1, param2, ...)**. Параметры подставляются на соответствующие места (см. примеры).

```python
s = 'potatos,eggs,milk,salt,tomatos'
t = '{0} is the first parameter in this pattern, while {1} is the second one.'
name = 'John'
age = 57

print(s)
print()
print(s.replace(',', ' '))
print(s.replace('to', 'TO'))
print()

print(t.format('param1', 'param2'))
print('My name is {0}. I am {1} years old.'.format(name, age))
# instead of print('My name is ' + name + 'I am ' + str(age) + 'years old.')
```

# Функции ord и chr.
У всех символов, которые могут присутствовать в строке есть свой номер. Чтобы получить этот номер, есть команда **ord**. В обратную сторону - **chr**

```python
print(ord('a'))
print(ord('A'))
for i in range(128):
    print(chr(i), end='')
```

Как мы видим, не все символы, различащиеся питоном, могут быть отображены стандартным шрифтом.
